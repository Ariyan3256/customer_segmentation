```python
x=df.iloc[:, 2:].values
x
```

```python
scale=MinMaxScaler()
features=scale.fit_transform(x)
features
```

```python
inertia=[]
for i in range(1, 16):
  k_means=KMeans(n_clusters=i)
  k_means=k_means.fit(features)
  inertia.append(k_means.inertia_)
```

```python
# Elbow graph
plt.figure(figsize=(20, 7))
plt.subplot(1, 2, 1)
plt.plot(range(1, 16), inertia, 'bo-')
plt.xlabel('No of clusters'), plt.ylabel('Inertia')
```

```python
#Kelbow visualizer
plt.subplot(1, 2, 2)
kmeans=KMeans()
visualize=KElbowVisualizer(kmeans, k=(1, 16))
visualize.fit(features)
plt.title("Elbow graph and elbow visualizer")
visualize.poof()
plt.show()
```

```python
silhouette_avg=[]
for i in range(2, 16):
  kmeans=KMeans(n_clusters=i)
  clusters_labels=kmeans.fit_predict(features)
  #Silhoutte score
  silhouette_avg.append(silhouette_score(features, clusters_labels))
```

```python
plt.figure(figsize=(10, 7))
plt.plot(range(2, 16), silhouette_avg, 'bX-')
plt.xlabel('Values of K')
plt.ylabel('Silhouette score')
plt.title('Silhouette analysis for optimal X')
plt.show()
```

Here we will take K value as 3 as per Elbow method

```python
model=KMeans(n_clusters=3)
model=model.fit(features)
```

```python
y_kn=model.predict(features)
centers=model.cluster_centers_
```

```python
df['Cluster']=pd.DataFrame(y_kn)
df.to_csv("Cluster_data", index=False)
```
